```python
import pandas as pd
# from pydantic_settings import BaseSettings # NEW
import ydata_profiling
from pandas_profiling import ProfileReport
```

```python
file_path = '/content/drive/MyDrive/HTF4_final/data/Retail_sale.xlsx'  # Đường dẫn đến file Excel của bạn
sheet1 = pd.read_excel(file_path, sheet_name='10.SE Regional Sales')
sheet2 = pd.read_excel(file_path, sheet_name='11.SE Regional Sales')
sheet3 = pd.read_excel(file_path, sheet_name='12.SE Regional Sales')
customer = pd.read_excel(file_path, sheet_name='Customer')
sale = pd.read_excel(file_path, sheet_name='Sale')
sale
```

```python
saleReport = ProfileReport(sale, title="Sale Report", explorative=True)
sheet1Report = ProfileReport(sheet1, title="Sale in October Report", explorative=True)
sheet2Report = ProfileReport(sheet2, title="Sale in November Report", explorative=True)
sheet3Report = ProfileReport(sheet3, title="Sale in December Report", explorative=True)
customerReport = ProfileReport(customer, title="Customer Report", explorative=True)
```
